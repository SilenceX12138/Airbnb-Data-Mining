```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
# CITY = 'beijing'
CITY = 'shanghai'
```

```python
room_df = pd.read_csv('data/{}/room.csv'.format(CITY))
```

```python
neighbourhood_group = room_df.groupby('neighbourhood')
```

```python
# 各行政区租金标准差
neighbourhood_group['price'].std().sort_values()[::-1].plot.bar(color='c')
plt.xlabel('行政区划')
plt.ylabel('租金标准差（元）')
plt.xticks(rotation=45)
```

```python
name_list = []
max_price_list = []
min_price_list = []
mean_price_list = []
for name, room in neighbourhood_group:
    name_list.append(name)
    max_price_list.append(room['price'].max())
    min_price_list.append(room['price'].min())
    mean_price_list.append(room['price'].mean())
# 最高日租金
plt.figure(0)
neighbourhood_group['price'].max().sort_values()[::-1].plot.bar(color='c')
plt.xlabel('行政区划')
plt.ylabel('最高日租金（元）')
plt.xticks(rotation=45)
# 最低日租金
plt.figure(1)
neighbourhood_group['price'].min().sort_values()[::-1].plot.bar(color='c')
plt.xlabel('行政区划')
plt.ylabel('最低日租金（元）')
plt.xticks(rotation=45)
# 平均日租金
plt.figure(2)
neighbourhood_group['price'].mean().sort_values()[::-1].plot.bar(color='c')
plt.xlabel('行政区划')
plt.ylabel('平均日租金（元）')
plt.xticks(rotation=45)
```

```python
# 房源价格分布
neighbourhood_group['price'].plot.kde(legend=True, figsize=(8, 5))
plt.xlim(-5000,27000)
plt.xlabel('日租金')
plt.ylabel('房源数量占比')
```
