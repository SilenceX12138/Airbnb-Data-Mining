```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
# CITY = 'beijing'
CITY = 'shanghai'
```

```python
best_room_df = pd.read_csv('data/{}/best_room.csv'.format(CITY))
```

```python
best_room_df.drop(['room_id','host_id','review_scores_rating'], inplace=True,axis=1)
# 
corr = best_room_df.corr()
sns.heatmap(corr, cmap="Blues", annot=True, fmt='.2f')
```

```python
best_room_df.plot.scatter(x='accommodates',y='price')
best_room_df.plot.scatter(x='bedrooms',y='price')
best_room_df.plot.scatter(x='beds',y='price')
best_room_df.plot.scatter(x='number_of_reviews',y='price')
```

```python
# 
best_room_df['neighbourhood'].value_counts().head(5).plot.bar(color='c')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=0)
```
