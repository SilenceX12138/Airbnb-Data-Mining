```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
# CITY = 'beijing'
CITY = 'shanghai'
```

```python
room_df = pd.read_csv('./data/{}/room.csv'.format(CITY))
```

```python
# 各区域房源数量占比
plt.figure(figsize=(15, 13))
room_df['neighbourhood'].value_counts().plot.pie(autopct='%.1f%%', pctdistance = 0.7,cmap=plt.get_cmap("tab20c"), textprops={'fontsize': 12, 'color': 'w'})
plt.legend(loc='upper left')
```

```python
# 房源数量前6
room_df['neighbourhood'].value_counts().plot.bar(width=0.9, color='c')
plt.xticks(rotation=45)
plt.xlabel('行政区划')
plt.ylabel('房源数量')
```
